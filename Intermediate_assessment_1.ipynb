**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
import pandas as pd
import numpy as np
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'], 'age' : [2.5 , 3 , 0.5 , np.nan , 5 , 2 , 4.5 , np.nan , 7 , 3] , 'visits' : [1 , 3 , 2 , 3 , 2 , 3 , 1 , 1 , 2 , 1] , 'priority' : ['yes' , 'yes' , 'no' , 'yes' , 'no' , 'no' , 'no' , 'yes' , 'no' , 'no']}
labels = ['a' , 'b' , 'c' , 'd' , 'e' , 'f' , 'g' , 'h' , 'i' , 'j']
df = pd.DataFrame(data , index = labels)
print (df)
````

**19.** Which airline ID is present maximum times in the dataset

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/ny-flights.csv')
airline_counts = df['airline_id'].value_counts()
max_airline = airline_counts.idxmax()
print(f"The maximum count is: {max_airline}")
````

**20.** Draw a plot between dep_delay and arr_delay

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/ny-flights.csv')
plt.figure(figsize=(20, 16))
sns.barplot(x='dep_delay', y='arr_delay', data=df)
plt.title('Bar Plot between Departure Delay and Arrival Delay')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.show()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20, 6))
sns.scatterplot(x='dep_delay', y='arr_delay', data=df)
plt.title('scatter Plot between Departure Delay and Arrival Delay')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.show()
````
